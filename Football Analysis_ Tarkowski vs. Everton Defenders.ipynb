```python
import pandas as pd
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

Read in data:

```python
df = pd.read_csv('Defending2.csv', sep=',')
df.head()
```

```python
for column in df.columns:
    print(column)
```

```python
params = list(df.columns)
params = params[3:]
params
```

```python
ranges = []

for x in params:
    a = min(df[params][x])
    a = a - (a*0.25)
    b = max(df[params][x])
    b = b + (b*0.25)
    ranges.append((a,b))

ranges
```

##### James Tarkowski vs. Yerry Mina

```python
print('James Tarkowski has played 9766 mins in the last 3 seasons')
print('Yerry Mina has played 5067 mins in the last 3 seasons')
```

```python
player_a = df.loc[df['Name ']=='James Tarkowski'].reset_index()
player_a = list(player_a.loc[0])
player_a = player_a[4:]


player_b = df.loc[df['Name ']=='Yerry  Mina'].reset_index()
player_b = list(player_b.loc[0])
player_b = player_b[4:]

values = [player_a, player_b]
print(values)
```

```python
title = dict(
    title_name='Yerry Mina',
    title_color='blue',
    subtitle_name='Everton',
    subtitle_color='blue',
    title_name_2='James Tarkowski',
    title_color_2='red',
    subtitle_name_2='Burnley',
    subtitle_color_2='red',
    title_fontsize=18,
    subtitle_fontsize=15
)

endnote = 'Sean Boyd\nData via FBREF / StatsBomb'

radar = Radar()

fig,ax = radar.plot_radar(ranges=ranges, params=params, values=values, 
                         radar_color=['red', 'blue'],
                         alphas=[0.75,0.6],
                         title=title,
                         endnote=endnote,
                         compare=True)

plt.savefig('TarkMina.eps', format='eps')
```
